Matplotlib vs. Seaborn
</span>




<div style="background:#243F36; padding:14px; border-radius:10px; margin-top:15px;">
  <p style="color:#B4E9D6; font-size:15px; margin:0;">

[Matplotlib](https://matplotlib.org/stable/plot_types/index.html)   is a comprehensive library for creating static, animated, and interactive visualizations in Python.<br>
[Seaborn](https://seaborn.pydata.org/examples/index.html) is a high level interface for drawing statistical graphics with Matplotlib.
  </p>
</div>

Matplotlib is a low-level visualization library that gives you <span style="color:#009688; font-weight:bold;">full control</span> over every element of a plot. Seaborn, built on top of Matplotlib, provides a higher-level interface with cleaner syntax and visually appealing default styles. While Matplotlib is more flexible and suitable for highly customized figures, Seaborn is ideal for <span style="color:#009688; font-weight:bold;">quick statistical</span> graphics with <span style="color:#009688; font-weight:bold;">minimal code</span>.

Plot Types
</span>


<span style="color:#009688; font-weight:bold;">Relational plots</span> :Show the relationship between two or more numeric variables (e.g., scatterplot, lineplot).  
<span style="color:#009688; font-weight:bold;">Distribution plots</span>  :Display the distribution and frequency of a dataset (e.g., histplot, kdeplot).  
<span style="color:#009688; font-weight:bold;">Categorical plots</span> :Compare values across different categories or groups (e.g., boxplot, barplot, stripplot).

<span style="color:#5EEAD4; font-size:36px;">
Figure-level vs. Axes-level functions
</span>

Axes-level functions plot data onto a <span style="color:#009688; font-weight:bold;">single matplotlib.pyplot.Axes</span>  object, which is the return value of the function.

In contrast, figure-level functions interface with matplotlib through a seaborn object, usually a <span style="color:#009688; font-weight:bold;">FacetGrid</span> , that manages the figure. Each module has a single figure-level function, which offers a unitary interface to its various axes-level functions.

you can use col= and row= only in figure_level functions.

```python
import numpy as np
import matplotlib.pyplot as plt
from itertools import cycle

from sklearn import svm, datasets
from sklearn.metrics import roc_curve, auc
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import label_binarize
from sklearn.multiclass import OneVsRestClassifier
from sklearn.metrics import roc_auc_score

# Import some data to play with
iris = datasets.load_iris()
X = iris.data
y = iris.target

# Binarize the output
y = label_binarize(y, classes=[0, 1, 2])
n_classes = y.shape[1]

# Add noisy features to make the problem harder
random_state = np.random.RandomState(0)
n_samples, n_features = X.shape
X = np.c_[X, random_state.randn(n_samples, 200 * n_features)]

# shuffle and split training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=0)

# Learn to predict each class against the other
classifier = OneVsRestClassifier(
    svm.SVC(kernel="linear", probability=True, random_state=random_state)
)
y_score = classifier.fit(X_train, y_train).decision_function(X_test)

# Compute ROC curve and ROC area for each class
fpr = dict()
tpr = dict()
roc_auc = dict()
for i in range(n_classes):
    fpr[i], tpr[i], _ = roc_curve(y_test[:, i], y_score[:, i])
    roc_auc[i] = auc(fpr[i], tpr[i])

# Compute micro-average ROC curve and ROC area
fpr["micro"], tpr["micro"], _ = roc_curve(y_test.ravel(), y_score.ravel())
roc_auc["micro"] = auc(fpr["micro"], tpr["micro"])
```

```python

sns.lineplot(x=fpr[2], y=tpr[2], errorbar=None)


sns.lineplot(x=[0,1], y=[0,1], linestyle="--")

plt.grid(True)

```
